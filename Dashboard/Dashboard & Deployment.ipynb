```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display
```

```python
df = pd.read_csv("data/02_detection_results.csv")
anomalies = pd.read_csv("data/03_anomalies.csv")
```

```python
fig = px.line(df, y="value", title="Anomaly Detection")
fig.add_scatter(x=anomalies.index, y=anomalies["value"], 
                mode="markers", name="Anomalies", marker=dict(color="red"))
fig.show()
```

```python
threshold_slider = widgets.IntSlider(value=50, min=0, max=120, step=5, description="Threshold:")
display(threshold_slider)

def filter_anomalies(threshold):
    filtered = anomalies[anomalies['value'] > threshold]
    display(filtered)

widgets.interactive(filter_anomalies, threshold=threshold_slider)
```
